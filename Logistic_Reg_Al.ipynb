```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df_diabete=pd.read_csv('diabete.csv')
```

```python
df_diabete.head()
```

```python
#checking null values
df_diabete.isnull().sum()
```

```python
#drop unneccessary columns
df_diabete.drop(columns=["Age","Insulin","SkinThickness"],axis=1,inplace=True)
```

```python
df_diabete
```

```python
df_diabete.drop(columns=["BMI"],axis=1,inplace=True)
```

```python
df_diabete
```

```python
df_diabete.drop(columns=["Pregnancies"],axis=1,inplace=True)
```

```python
df_diabete.head()
```

```python
#creating independent and dependent features
x=df_diabete.iloc[:,:-1]
y=df_diabete.iloc[:,-1]
```

```python
x
```

```python
#training testing spliting data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train.shape
```

```python
#Feature scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
x_train
```

```python
#training the model
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
#testing the model
#predicting the test model
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
#confusion matrics accuracy
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
